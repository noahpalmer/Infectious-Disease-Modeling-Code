```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

cbar = 0.45;
popfactor = 4;
b1 = cbar*popfactor*1; b2 = cbar*popfactor*2; b3= cbar*popfactor*3; b4 = cbar*popfactor*4;

T = 12; dt = 0.01
steps = int(T/dt)
t = np.linspace(0, T, steps)

S1 = np.zeros(steps); I1 = np.zeros(steps); R1 = np.zeros(steps)
S2 = np.zeros(steps); I2 = np.zeros(steps); R2 = np.zeros(steps)
S3 = np.zeros(steps); I3 = np.zeros(steps); R3 = np.zeros(steps)
S4 = np.zeros(steps); I4 = np.zeros(steps); R4 = np.zeros(steps)
p_avg = np.zeros(steps)

S1[0] = 0.999/4; S2[0] = 0.999/4; S3[0] = 0.999/4; S4[0] = 0.999/4;
I1[0] = 0.001/4; I2[0] = 0.001/4; I3[0] = 0.001/4; I4[0] = 0.001/4;
R1[0] = 0; R2[0] = 0; R3[0] =0; R4[0] = 0;

p_avg[0] = (1*S1[0]+2*S2[0]+3*S3[0]+4*S4[0]) / (S1[0]+S2[0]+S3[0]+S4[0])


for i in range(steps-1):

    sumI= I1[i]+I2[i]+I3[i]+I4[i];
    
    dS1_dt = -b1*S1[i]*(sumI)
    dS2_dt = -b2*S2[i]*(sumI)
    dS3_dt = -b3*S3[i]*(sumI)
    dS4_dt = -b4*S4[i]*(sumI)
    
    dI1_dt = b1*S1[i]*(sumI)-3*I1[i]
    dI2_dt = b2*S2[i]*(sumI)-3*I2[i];
    dI3_dt = b3*S3[i]*(sumI)-3*I3[i];
    dI4_dt = b4*S4[i]*(sumI)-3*I4[i];
    
    dR1_dt = 3*I1[i]
    dR2_dt = 3*I2[i]
    dR3_dt = 3*I3[i]
    dR4_dt = 3*I4[i]
  

    S1[i+1] = S1[i] + dS1_dt*dt
    I1[i+1] = I1[i] + dI1_dt*dt
    R1[i+1] = R1[i] + dR1_dt*dt
    S2[i+1] = S2[i] + dS2_dt*dt
    I2[i+1] = I2[i] + dI2_dt*dt
    R2[i+1] = R2[i] + dR2_dt*dt
    S3[i+1] = S3[i] + dS3_dt*dt
    I3[i+1] = I3[i] + dI3_dt*dt
    R3[i+1] = R3[i] + dR3_dt*dt
    S4[i+1] = S4[i] + dS4_dt*dt
    I4[i+1] = I4[i] + dI4_dt*dt
    R4[i+1] = R4[i] + dR4_dt*dt

    
    p_avg[i+1] = (1*S1[i+1]+2*S2[i+1]+3*S3[i+1]+4*S4[i+1]) / (S1[i+1]+S2[i+1]+S3[i+1]+S4[i+1])



colorsr = plt.cm.Reds(np.linspace(0.4, 1, 4))  
colorsb = plt.cm.Blues(np.linspace(0.4, 1, 4)) 

plt.figure(figsize=(8,5))
plt.plot(t, I1, color=colorsr[0], label="Group 1 Infected")
plt.plot(t, I2, color=colorsr[1], label="Group 2 Infected")
plt.plot(t, I3, color=colorsr[2], label="Group 3 Infected")
plt.plot(t, I4, color=colorsr[3], label="Group 4 Infected")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(t, S1, color=colorsb[0], label="Group 1 Susceptible")
plt.plot(t, S2, color=colorsb[1], label="Group 2 Susceptible")
plt.plot(t, S3, color=colorsb[2], label="Group 3 Susceptible")
plt.plot(t, S4, color=colorsb[3], label="Group 4 Susceptible")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(t, p_avg, color="black", linewidth=2, label="Average Relative Susceptibility")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```
