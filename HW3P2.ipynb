```python
import numpy as np
import time
from scipy.stats import nbinom

def branching(R0, k, G=100, M=100000): #Large G plus large M is making this never terminate...
    nchainend = 0
    mean = R0
    variance = mean+(mean**2)/k
    p = mean/variance
    n = mean**2/(variance-mean)
    for i in range(M):
        ninfected = 1  
        for j in range(G):
                offspring = nbinom.rvs(n=n, p=p, size=ninfected)
                ninfected = np.sum(offspring)
                if ninfected == 0:
                    nchainend += 1
                    break
                if ninfected>=200:
                    break
                #Exit condition on number of children 200, 100 generation
    q = nchainend/M
    return q
kvals = [0.1,0.5,1,5,10]
results = {}
start_time = time.perf_counter()
for k in kvals:
    results[k] = branching(R0=3, k=k)

end_time = time.perf_counter()
elapsed_time = end_time - start_time
print(f"branching() executed in {elapsed_time:.6f} seconds")
```
