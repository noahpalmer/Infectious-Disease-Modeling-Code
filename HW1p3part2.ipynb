```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
beta = 1; gamma = 0.5; 

T = 100; dt = 0.05
steps = int(T/dt)
t = np.linspace(0, T, steps)

S = np.zeros(steps); I = np.zeros(steps); R = np.zeros(steps)

# ICs
N = 1000
I[0] = 1; S[0] = 999; R[0] = 0


for i in range(steps-1):
    
    dS_dt = - beta*S[i]*I[i]/N 
    dI_dt = beta*S[i]*I[i]/N - gamma*I[i]
    dR_dt = gamma*I[i] 

    S[i+1] = S[i] + dS_dt*dt
    I[i+1] = I[i] + dI_dt*dt
    R[i+1] = R[i] + dR_dt*dt


plt.figure(figsize=(10, 6))
plt.plot(t, S, color = 'blue', label='Susceptible')
plt.plot(t, I, color = 'red', label='Infected')
plt.plot(t, R, color = 'black', label='Recovered')
plt.axhline(0.797 * N, color='green', linestyle=':', label='Predicted amount of recovered individuals')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Noah Palmer: SIR Simulation')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


    
```
