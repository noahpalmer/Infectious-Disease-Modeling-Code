```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
beta = 3; gamma = 2; R0 = beta/gamma

T = 25; dt = 0.5 #Each plot made by modifying step size here
steps = int(T/dt)
t = np.linspace(0, T, steps)
t2 = np.linspace(0,T,2000)

S = np.zeros(steps); I = np.zeros(steps); 


I[0] = 0.01; S[0] = 0.99;

analytic = (1-1/R0)/(1+((1-1/R0-I[0])/I[0])*np.exp(-(beta-gamma)*t2))

steps_taken = 0 

for i in range(steps-1):
    
    dS_dt = - beta*S[i]*I[i]+gamma*I[i]
    dI_dt = beta*S[i]*I[i]- gamma*I[i]

    S[i+1] = S[i] + dS_dt*dt
    I[i+1] = I[i] + dI_dt*dt

plt.figure(figsize=(10, 6))
plt.plot(t2, analytic, color = 'black', linestyle='--', label='Analytical')
plt.plot(t, I, color = 'red', label='Forward Euler')
plt.xlabel('Time')
plt.ylabel('Infections')
plt.title('Simulation of normalized SIS')
plt.legend()
plt.grid(True)
plt.ylim(0, 0.5)
plt.tight_layout()
plt.show()



    

```
