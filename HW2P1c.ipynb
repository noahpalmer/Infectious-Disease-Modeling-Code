```python
import numpy as np
import matplotlib.pyplot as plt

T=25; beta =3; gamma=2; I0=0.01; S0=0.99


def siserror(dt_values, T, beta, gamma, I0, S0):
    
    R0 = beta / gamma
    errs = {}


    for dt in dt_values:
        steps = int(T / dt)
        t = np.linspace(0, T, steps)
        S = np.zeros(steps)
        I = np.zeros(steps)
        S[0] = S0
        I[0] = I0
        analytic = (1 - 1 / R0) / (1 + ((1 - 1 / R0 - I0) / I0) * np.exp(-(beta - gamma) * t))

        for i in range(steps - 1):
            dS_dt = - beta*S[i]*I[i]+gamma*I[i]
            dI_dt = beta*S[i]*I[i]- gamma*I[i]
            S[i+1] = S[i] + dS_dt*dt
            I[i+1] = I[i] + dI_dt*dt
            
        error = max_error = np.max(np.abs(I - analytic))
        errs[dt] = error

    return errs

dt_values = [2,1,1/2,1/4,1/8,1/16,1/32]
siserror(dt_values, T, beta, gamma, I0, S0)



    

```
