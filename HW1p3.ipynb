```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def f(r):
    return r

def g(r, R0):
    return 1-np.exp(-R0*r)

def h(r,R0): 
    return f(r) - g(r,R0)

R0vals = [0.9, 1.0, 1.1, 1.2]

r = np.linspace(0, 1, 1000)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, R0 in enumerate(R0vals):

    row, col = divmod(i, 2)
    ax = axes[row, col]
    rguess = 0.6
    root = fsolve(h, rguess, args = (R0))[0]

    fsoln = f(root)

    ax.plot(r, f(r), 'k-', label='$f(r_\\infty)$')
    ax.plot(r, g(r, R0), 'r-', label='$g(r_\\infty)$')
    ax.scatter(root, fsoln, color='blue', s=50, label=f' $r_\\infty$ ≈ {root:.3f}')

     # Titles and labels
    ax.set_title(f'$R_0 = {R0}$')
    ax.set_xlabel('$r_\\infty$')
    ax.set_ylabel('Function Value')
    ax.legend()
```
