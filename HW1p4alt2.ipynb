```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
beta = 1; 

T = 2000; dt = 0.1
steps = int(T/dt)
t = np.linspace(0, T, steps)

S = np.zeros(steps); I = np.zeros(steps); R = np.zeros(steps)

N = 10**(6)

gammavals = [ 0.998, 0.999,0.999999, 1.001,1.01, 1.1]

plt.figure(figsize=(10, 6))

for gamma in gammavals:
    S = np.zeros(steps)
    I = np.zeros(steps)
    R = np.zeros(steps)

    I[0] = 1
    S[0] = N - I[0]
    R[0] = 0
    R0 = 1/gamma
    
    for i in range(steps - 1):
        dS_dt = - beta * S[i] * I[i] / N
        dI_dt = beta * S[i] * I[i] / N - gamma * I[i]
        dR_dt = gamma * I[i] 

        S[i + 1] = S[i] + dS_dt * dt
        I[i + 1] = I[i] + dI_dt * dt
        R[i + 1] = R[i] + dR_dt * dt


    plt.plot(t, I, label=f'Infected ($R_0$ = {R0})')

plt.xlabel('Time')
plt.ylabel('Number of infected individuals')
plt.title('Noah Palmer: Infected indiduals for different $R_0$ values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


    
```
